```python
import requests
import pandas as pd

url = "https://api.sofascore.com/api/v1/unique-tournament/325/season/58766/standings/total"

# Faz a solicitação GET para a API
response = requests.get(url)

# Verifica se a solicitação foi bem sucedida (código de status 200)
if response.status_code == 200:
    # Extrai os dados JSON da resposta
    data = response.json()
    
    # Obtém a lista de times
    standings = data["standings"][0]["rows"]
    
    # Cria uma lista de dicionários com os dados de cada time
    teams_data = []
    for team in standings:
        team_data = {
            "team_names": team["team"]["name"],
            "Position": team["position"],
            "Points": team["points"],
            "MatchesPlayed": team["matches"],
            "Wins": team["wins"],
            "Draws": team["draws"],
            "Losses": team["losses"],
            "GoalsFor": team["scoresFor"],  # Corrigido para a chave correta
            "GoalsAgainst": team["scoresAgainst"]  # Corrigido para a chave correta
        }
        teams_data.append(team_data)
    
    # Cria um DataFrame a partir da lista de dicionários
    tabela = pd.DataFrame(teams_data)
    
    # Exibe o DataFrame
    print(tabela)
else:
    print("Falha na solicitação. Código de status:", response.status_code)

# Salva o DataFrame em um arquivo CSV
tabela.to_csv('tabela.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# URL da página
url = "https://www.transfermarkt.com.br/campeonato-brasileiro-serie-a/startseite/wettbewerb/BRA1"

# Headers para simular uma requisição de um navegador web
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

response = requests.get(url, headers=headers)

# Criar um objeto BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar a tabela que contém os dados dos clubes
table = soup.find('table', {'class': 'items'})

# Lista para armazenar os dados
data = []

# Iterar sobre cada linha da tabela exceto o cabeçalho
for row in table.find_all('tr')[1:]:
    # Extrair as células de cada linha
    cols = row.find_all('td')
    if cols:
        # Extrair os dados necessários
        clube = cols[1].text.strip()  # Índice ajustado para clube
        media_idade = cols[3].text.strip()  # Índice ajustado para média de idade
        valor_mercado_total = cols[6].text.strip()  # Índice ajustado para valor de mercado total
        data.append([clube, media_idade, valor_mercado_total])

```

```python
import pandas as pd

# Convertendo para DataFrame
df = pd.DataFrame(data)

# Removendo a primeira linha
df = df.drop(0)

# Resetando os índices
df = df.reset_index(drop=True)

# Renomeando as colunas
df = df.rename(columns={0: 'team_names', 1: 'avg_age', 2: 'squad_value_euro'})

# Removendo caracteres não numéricos e convertendo para float
df['squad_value_euro'] = df['squad_value_euro'].str.replace('mi. €', '').str.replace(',', '.').astype(float)

# Multiplicando por milhões
df['squad_value_euro'] *= 1000000

# Dicionário de mapeamento para os nomes dos times
# Dicionário de mapeamento para os nomes dos times
team_mapping = {
    'Athletico Paranaense': 'Athletico',
    'EC Bahia': 'Bahia',
    'Botafogo FR': 'Botafogo',
    'Atlético Mineiro': 'Atlético Mineiro',
    'RB Bragantino': 'Red Bull Bragantino',
    'SE Palmeiras': 'Palmeiras',
    'CR Flamengo': 'Flamengo',
    'São Paulo FC': 'São Paulo',
    'SC Internacional': 'Internacional',
    'Cruzeiro EC': 'Cruzeiro',
    'Grêmio FBPA': 'Grêmio',
    'Fortaleza EC': 'Fortaleza',
    'Criciúma EC': 'Criciúma',
    'SC Corinthians': 'Corinthians',
    'EC Juventude': 'Juventude',
    'Fluminense FC': 'Fluminense',
    'CR Vasco da Gama': 'Vasco da Gama',
    'EC Vitória': 'Vitória',
    'Atlético Goianiense': 'Atlético Goianiense',
    'Cuiabá EC': 'Cuiabá'
}

# Substituindo os nomes dos times usando o mapeamento
df['team_names'] = df['team_names'].replace(team_mapping)

print(df)

# Salvando o DataFrame em um arquivo CSV
df.to_csv('valores.csv', index=False)

# Nome_do_arquivo.csv é o nome do arquivo que você deseja salvar
# index=False evita que o pandas inclua o índice do DataFrame no arquivo CSV

print("DataFrame salvo em CSV com sucesso!")
```
